```python
#56 . ARIMA model for Netflix
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(df["Netflix: (India)"], order=(1, 1, 1))
model_fit = model.fit()
print(model_fit.summary())
```

```python
# 57 . Forecasting with ARIMA
forecast = model_fit.forecast(steps=4)
plt.figure(figsize=(12, 6))
plt.plot(df["Week"], df["Netflix: (India)"], label="Netflix: (India)", color='red', marker='o')
plt.plot(pd.date_range(start=df["Week"].iloc[-1], periods=5, freq='W')[1:], forecast, label="Forecast", color='blue', linestyle='--')
plt.title("Netflix Forecast with ARIMA")
```

```python
# 58 . Seasonal ARIMA model
from statsmodels.tsa.statespace.sarimax import SARIMAX
seasonal_model = SARIMAX(df["Netflix: (India)"], order=(1, 1, 1), seasonal_order=(1, 1, 1, 4))
seasonal_model_fit = seasonal_model.fit()
print(seasonal_model_fit.summary())
```

```python
# 59 . Forecasting with Seasonal ARIMA
seasonal_forecast = seasonal_model_fit.forecast(steps=4)
plt.figure(figsize=(12, 6))
plt.plot(df["Week"], df["Netflix: (India)"], label="Netflix: (India)", color='red', marker='o')
plt.plot(pd.date_range(start=df["Week"].iloc[-1], periods=5, freq='W')[1:], seasonal_forecast, label="Seasonal Forecast", color='blue', linestyle='--')
plt.title("Netflix Seasonal Forecast with SARIMA")
```

```python
import pandas as pd
import plotly.express as px

# 1. Convert Week to datetime
df["Week"] = pd.to_datetime(df["Week"])
df["Year"] = df["Week"].dt.year

# 2. Identify all OTT platform columns (exclude 'Week' and other non-numeric)
ott_columns = [col for col in df.columns if "(India)" in col]

# 3. Melt the DataFrame to long format
long_df = df.melt(id_vars=["Week", "Year"], value_vars=ott_columns,
                  var_name="Platform", value_name="Trend")

# Optional: Clean platform names
long_df["Platform"] = long_df["Platform"].str.replace(": \(India\)", "", regex=True)

# 4. Group by Platform and Year (you can also add Month if needed)
grouped_df = long_df.groupby(["Platform", "Year"]).sum(numeric_only=True).reset_index()

# 5. Create Sunburst Chart
fig = px.sunburst(
    grouped_df,
    path=["Platform", "Year"],
    values="Trend",
    title="OTT Platform Trends in India by Year",
    template="plotly_dark",
    color="Trend",
)

fig.show()
```
